```python
import pandas as pd
# 读入数据集
titanic = pd.read_csv("channing.csv")
```

```python
# 打印csv中的数据
print(titanic)
```

```python
# 查看前12行数据
print(titanic.head(12))
```

```python
# 检查每列的数据类型
print(titanic.dtypes)
```

```python
# 导出为excel表
titanic.to_excel("channing.xlsx",sheet_name="channing", index=False)
```

```python
# 查看每列的信息
titanic.info()
```

```python
# 查看居民的性别以及在Channing House 度过的时间长度
sex_time = titanic[["sex","time"]]
print(sex_time)
```

```python
# 列出在Channing House 度过的时间长度超过130的居民
print(titanic[titanic["time"] > 130])
```

```python
# 列出死亡时的年龄等于950月的居民
print(titanic[titanic["exit"].isin([900,950])])
```

```python
# 查看DataFrame中哪些条目不是NA
print(titanic.notna())
```

```python
# 查看第9行到第25行、第3列到第6列的数据
titanic.iloc[9:25, 3:6]
```

```python
# 使用matplotlib生成图表
import matplotlib.pyplot as plt
titanic.plot()
plt.show()
```

```python
# 生成箱线图
titanic.plot.box()
plt.show()
```

```python
# 只查看居民在 Channing House 度过的时间长度（以月为单位）
titanic["time"].plot()
plt.show()
```

```python
# 生成居民存活时间的散点图
titanic.plot.scatter(x="num",y="time",alpha=0.5)
plt.show()
```

```python
[
    method_name
    for method_name in dir(titanic.plot)
    if not method_name.startswith("_")
]
titanic.plot.box()
plt.show()
```

```python
# 每一列形成一个单独的子图
axs = titanic.plot.area(figsize=(12, 4), subplots=True)
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12, 4))        # 创建一个新的matplotlib图
titanic.plot.area(ax=axs)                       # 将图放在准备好的Axes上
axs.set_ylabel("NO$_2$ concentration")          # 定制图表
fig.savefig("no2_concentrations.png")           # 使用现有方法保存图形
plt.show()                                      
```
